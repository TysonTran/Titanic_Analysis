```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2)
%matplotlib inline
%config InlineBackend.figure_format = 'retina'  # Higher resolution figures
```

```python
df = pd.read_csv('train.csv')
df
```

Droping Columns that add no value to the data

1. Passenger Id - This variable is not very useful because we do not know if it is the order that they boarded the ship, when they bought their ticket, or something else.

2. Cabin - This does not add a lot of value because there are lots of passengers who share rooms with others onboard, there are a lot of NaN values, and we already know what part of the ship the passenger is on by their Pclass

3. Ticket - This variable does not look like it has a pattern.

```python
df = df.drop(labels=['PassengerId', 'Cabin', 'Ticket'], axis=1)
df
```

```python
df.info()
```

```python
df.isna().sum() #Data will be imputated later on in the analysis
```

### What is the survival of Males versus Females on the Titanic?


```python
df['Sex'].value_counts()
```

There are 577 males on the ship and 314 females on the ship with a total of 891 individuals on the Titanic

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['Sex'].value_counts().plot(kind='bar', color=['black','green'], title='Count of Male vs Female Onboard', ax=ax[0])
(df['Sex'].value_counts()/df['Sex'].value_counts().sum()).plot(kind='bar', color=['black','green'], ax=ax[1], title='Proportion of Male vs. Female')
plt.show()
```

As we can see from the figure above, there is almost double the number of males on the ship than females.


```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Proportion of Survived by Gender')
sns.countplot(x='Sex',hue='Survived',data=df,ax=ax[1])
ax[1].set_title('Count of Survived and Dead by Gender')
plt.show()
```

There was almost double the number of males on the ship than females, but the rate of survival for females was much higher than males. Over 70% of females survived while less than 20% of the male passengers survived

### Does the ticket class affect the survival rate?

```python
f,ax=plt.subplots(1,3,figsize=(18,8))
df['Pclass'].value_counts().sort_index().plot.bar(ax=ax[0])
ax[0].set_title('Count of Passenger in Each Ticket Class')
df[['Survived', 'Pclass']].groupby('Pclass').mean().plot.bar(ax=ax[1])
ax[1].set_title('Proportion of Survived in Each Ticket Class')
sns.countplot(x='Pclass',hue='Survived',data=df,ax=ax[2])
ax[2].set_title('Count of Survived and Dead by Ticket Class')
plt.show()
```
